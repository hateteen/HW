## Условие №3

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.

Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
plt.figure(figsize =(8, 6))
sns.lineplot(x = df1['price'], y = df1['sqft_living'])
plt.title('Living area and price correlation')
plt.xlabel('price')
plt.ylabel('living area')
;
```

Вывод: стоимость недвижимости линейно растет с ростом жилой площади, при этом количество недвижимости падает.

```python
plt.figure(figsize =(8, 6))
sns.boxplot(y = df1['price'], x = df1['floors'])
plt.title('Floors amount and price correlation')
plt.xlabel('floors')
plt.ylabel('price');
```

Вывод: Дороже всего стоят дома с этажностью 2.5. При этом и медиана у них самая большая. Дома с этажностью 3.5 имеют самую большую минимальную стоимость. Рост цен идет до этажности 2.5. Далее ввиду малого количества домов большой этажности и спроса цены падают. 

```python
plt.figure(figsize =(8, 6))
sns.boxplot(y = df1['price'], x = df1['waterfront'])
plt.title('Waterfront availability and price correlation')
plt.xlabel('waterfront')
plt.ylabel('price');
```

Вывод: Дома, расположенные у набережной имеют более высоку цену, но малое количество выбросов. При этом цены на дома без набережной более низкие, но имеют большое количество выбросов. Возможно это результат большоего количеста объектов недвижимости.

```python
plt.figure(figsize =(8, 6))
sns.boxplot(y = df1['price'], x = df1['view'])
plt.title('View quality and price correlation')
plt.xlabel('view')
plt.ylabel('price');
```

Вывод: Чем лучше вид, тем дороже недвижимость.

```python
plt.figure(figsize =(8, 6))
sns.lineplot(x = df1['yr_built'], y = df1['price'])
plt.title('Build year and price correlation')
plt.xlabel('price')
plt.ylabel('year');
```

Вывод: Цены на джома до 1930 года - самые высокие. Цены на недвижимость с 1930 до 1980 года низкие. Цены на дома моложе 1980 года - высокие. Можно рассмотреть несколько причин таких результатов: 
* Старые дома ценятся за историю и архитектуру. Имеют хорошее расположение. А так же в них скорее всего произведена реновация.
* Новые дома имеют современный дизайн и коммуникации. Не требуют ремонта.
* Остальные возможно находятся в плохом состоянии и требуют вложений. Возможно имеют плохое расположение.
